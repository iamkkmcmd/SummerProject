### Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data: bank_details
bank_details = pd.read_csv('./../data/bank_details.csv', encoding = 'latin')
bank_details['Name of the Bank'] = bank_details['Name of the Bank'].apply(lambda x: x.upper())
bank_details.head()
```

```python
# Read the data: bank_firm_connection
bank_firm_connection = pd.read_excel('./../data/bank_firm_connection.xlsx')
bank_firm_connection.head()
```

Problem: Name of the Bank and banker_name are different. We need to fix it.

```python
# Finding unique name of the bank taken into caonsideration
required_bank = np.sort(bank_details['Name of the Bank'].unique())
print(required_bank)
```

```python
# Finding unique banker name 
given_bank = np.sort(bank_firm_connection['banker_name'].unique())
print(given_bank)
```

```python
# Let's make a function to count number of index wise mached characters
def count_match(word, target):
    length = min(len(word), len(target))
    count = 0
    for i in range(length):
        if word[i] == target[i]:
            count += 1
    return count
```

```python
# filtered out the banker_name
filtered_bank = []
for bank in required_bank:
    filtered_bank.append(given_bank[np.argmax([count_match(bank, word) for word in given_bank])])
```

```python
# combine and making required dataframe
df = pd.DataFrame({'required_bank':required_bank, 'filtered_bank':filtered_bank})
df
```

```python
# filter the bank-firm connection dataframe
bank_firm_connection = bank_firm_connection[bank_firm_connection['banker_name'].isin(filtered_bank)]
bank_firm_connection.sort_values(by = ['banker_name', 'company_name'], inplace = True, ignore_index = True)
```

```python
# print the dataframe
bank_firm_connection
```

```python
# save the file
bank_firm_connection.to_csv('./../data/bank_firm_connection_filter.csv', index = False)
```
